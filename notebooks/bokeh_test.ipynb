```python
import hail as hl
hl.init()
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
mt = hl.read_matrix_table('/home/olavur/experiments/2020-11-13_fargen1_exome_analysis/data/mt/ld_pruned_diallelic.mt')
```

Subset loci.

```python
mt_small = mt.sample_rows(0.001)
print(mt_small.count())
```

Get the positions (integers) and collect them to a list.

```python
positions = mt_small.locus.position.collect()
idx = range(len(positions))
```

Plot the positions and the indexes of the positions.

```python
# create a new plot with a title and axis labels
p = figure(title="", x_axis_label='idx', y_axis_label='position')

# add a line renderer with legend and line thickness
p.line(idx, positions, line_width=2)

# show the results
show(p)
```
